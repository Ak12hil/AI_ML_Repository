```python
from bs4 import BeautifulSoup
with open("scraped_data/data1.html") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
soup.select_one("span.country-population")
```

```python
#extract country names in the html .. h3 tags with classname country_name

all_h3=soup.find_all("h3")
names=[]
population=[]
all_data=[]
for h3 in all_h3:
    all_data.append([h3.get_text(strip=True),h3.find_next("div").select("span.country-population")[0].get_text(strip=True)])
    names.append(h3.get_text(strip=True))
    population.append(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
#print(population)
```

```python
import pandas as pd

df=pd.DataFrame(all_data,columns=["Names","Population"])
df.to_csv("cleaned_data/data1.csv",index=False)
```
