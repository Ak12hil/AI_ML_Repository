```python
# Real data
oscar_movies=[
    "The dark knight",
    "The hunter locker",
    "The king's speech",
    "The artist"
]

years=[2008,2009,2010,2011]
oscar_revenue=[1005,170,427,133]

oscar_plot=plt.plot(years,oscar_revenue)
plt.title("Revenue of oscar movies in the Respective Years")
plt.xlabel("Years")
plt.ylabel("Revenue($M)")
```

```python
non_oscar_movies=[
    "Slumdog millionaire",
    "Avatar",
    "Inception",
    "Hugo"
]

non_years=[2008,2009,2010,2011]
non_oscar_revenue=[378,2788,829,185]

plt.title("Revenue of oscar movies VS non-oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue($M)")
plt.plot(non_years,non_oscar_revenue,'o--r',label="non-oscar movies")
plt.plot(years,oscar_revenue,'>--y',label="oscar movies")
plt.legend()
```

```python
non_oscar_movies=[
    "Slumdog millionaire",
    "Avatar",
    "Inception",
    "Hugo"
]

non_years=[2008,2009,2010,2011]
non_oscar_revenue=[378,2788,829,185]

plt.title("Bar chart for Revenue of oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue($M)")
plt.bar(non_years,non_oscar_revenue,label="non-oscar movies")
for i in range(0,len(years)):
    plt.text(non_years[i],non_oscar_revenue[i]+10,str(non_oscar_revenue[i]),ha="center")
plt.legend()

plt.ylim(0,max(non_oscar_revenue)+1000)
```

```python
import numpy as np
non_oscar_movies=[
    "Slumdog millionaire",
    "Avatar",
    "Inception",
    "Hugo"
]

non_years=[2008,2009,2010,2011]
non_oscar_revenue=[378,2788,829,185]

plt.title("Revenue of oscar movies VS non-oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue($M)")
x=np.arange(len(years))
width=0.4
plt.bar(x-width/2,non_oscar_revenue,width,label="non-oscar movies")
plt.bar(x+width/2,oscar_revenue,width,label="oscar movies")
plt.legend()
```

```python
# Assignment
# compare temperature of cities over a week using subplots and loops

days=["Mon","Tue", "Wed","Thu","Fri"]
cities=["New York","London","Delhi","Tokyo"]

temperatures=[
    [22,23,21,24,25], #New York
    [18,19,17,20,21], #London
    [30,32,31,33,34], #Delhi
    [25,26,24,27,28] #Tokyo
]
fig,ax=plt.subplots(2,2)

for i in range(0,int(len(cities)/2)):
    for j in range(0,int(len(cities)/2)):
        ax[i][j].plot(days,temperatures[i+j],marker='o')
        ax[i][j].set_title(cities[i+j])
        ax[i][j].grid()

# ax[0][0].plot(days,temperatures[0])
# ax[0][0].set_title(cities[0])

# ax[0][1].plot(days,temperatures[1])
# ax[0][1].set_title(cities[1])

# ax[1][0].plot(days,temperatures[2])
# ax[1][0].set_title(cities[2])

# ax[1][1].plot(days,temperatures[3])
# ax[1][1].set_title(cities[3])

fig.tight_layout()
```
