```python
# melt and pivot
data=pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax":[10,20,30,40]
})
melted_data=data.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit","tax"],
    var_name="metrics",
    value_name="value"
)
melted_data
```

```python
#pivot
pivoted_data=melted_data.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
pivoted_data
```
