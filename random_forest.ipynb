# Random forest for classification

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
features=["pclass","sex","fare","embarked","age"]
target=["survived"]

#fill missing values
from sklearn.impute import SimpleImputer
imputer_median=SimpleImputer(strategy="median")
imp_fre=SimpleImputer(strategy="most_frequent")

titanic[["age"]]=imputer_median.fit_transform(titanic[["age"]])
titanic[["embarked"]]=imp_fre.fit_transform(titanic[["embarked"]])

```

```python
# encoding
# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

titanic["sex"]=le.fit_transform(titanic["sex"])
titanic["embarked"]=le.fit_transform(titanic["embarked"])
```

```python
X=titanic[features]
y=titanic["survived"]
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# descision tree model -no pruning
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()

model.fit(Xtrain,ytrain)

ypred=model.predict(Xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score: ",accuracy_score(ytest,ypred))
print("accuracy score: ",accuracy_score(ytrain,model.predict(Xtrain)))
# classic case of overfitting
```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(
    n_estimators=201,
    oob_score=True
)

rf.fit(Xtrain,ytrain)
ypred=rf.predict(Xtest)

print("OOB score: ",rf.oob_score_)
```
