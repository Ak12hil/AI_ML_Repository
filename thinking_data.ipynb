```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store_data.json")
print(data)
```

```python
#clean the data
def clean_data(data):
    text_to_num={"one":1, "two":2, "three":3 , "four":4, "five":5}
    cleaned_data=[]
    unique_users=set()
    for record in data:
        # clean the inconsitency in the rating
        raw_rating=str(record["rating"]).strip().lower()
        if raw_rating in text_to_num.keys():
            record["rating"]=text_to_num[raw_rating]

        #handle missing values ex:age
        if record.get("age")==None:
            record["age"]=None

        #deduplication
        if record["name"].strip() in unique_users:
            continue
        unique_users.add(record["name"])
        cleaned_data.append(record)
    return cleaned_data
            
```

```python
cleaned_data=clean_data(data)
print(cleaned_data)
```

```python
# Meaningful ingsights
def avg_rating(data):
    sum=0;
    for record in data:
            sum=sum+float(record["rating"])
    return sum/len(data)

def poor_rating_percentage(data):
    total_poor_rating=0
    for record in data:
        if float(record["rating"])<3.0:
            total_poor_rating=total_poor_rating+1
    return (total_poor_rating * 100)/len(data)

print(f"Average rating is {avg_rating(cleaned_data)}")
print(f"Poor rating percentage is {poor_rating_percentage(cleaned_data)}%")
```

```python
#recommendatation feature
def get_brand_recommendations(data):
    recommendations=[]
    for record in data:
        current_recommendation={}
        current_recommendation["name"]=record["name"]
        if float(record["rating"])>=4:
            current_recommendation["brand"]="Apple"
        else:
            current_recommendation["brand"]="Samsung"
        
        recommendations.append(current_recommendation)    
            
    return recommendations

print(get_brand_recommendations(cleaned_data))
    
```
