```python
# From Quotes , extract all quotes and authors from all pages with the life tag
#https://quotes.toscrape.com/
```

```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url="https://quotes.toscrape.com/"
response=requests.get(url)

#write the data to file
with open("scraped_data/quotesData.html","w") as f:
    f.write(response.text)
```

```python
with open("scraped_data/quotesData.html") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")
```

```python
all_tagElements=soup.select("a.tag")
quote_info=[]
for ele in all_tagElements:
    if ele.get_text(strip=True)=="life":
        quote=ele.find_parent().find_parent().select_one("span.text")
        if quote!=None:
            quote=quote.get_text()
        author=ele.find_parent().find_parent().select_one("small.author")
        if author!=None:
            author=author.get_text()
        quote_info.append([quote,author])
        # print()
        # break
```

```python
import pandas as pd
quotes=pd.DataFrame(quote_info,columns=["Quotes","Author"])
quotes
```
