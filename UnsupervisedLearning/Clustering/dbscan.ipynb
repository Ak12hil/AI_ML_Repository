```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris=load_iris()
X=iris.data
scaler=StandardScaler()
```

```python
Xscaled=scaler.fit_transform(X)
```

```python
sns.scatterplot(x=Xscaled[:,0],y=Xscaled[:,2],c=iris.target)
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(
    eps=0.8,
    min_samples=5
)
labels=model.fit_predict(Xscaled)
```

```python
sns.scatterplot(x=Xscaled[:,0],y=Xscaled[:,2],c=labels)
```

# for non linear data

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(
    random_state=42,
    n_samples=300,
    noise=0.05
)
```

```python
Xscaled=scaler.fit_transform(X)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
model2=DBSCAN(
    eps=0.5,
    min_samples=5
)
labels2=model2.fit_predict(Xscaled)
```

```python
sns.scatterplot(x=Xscaled[:,0],y=Xscaled[:,1],c=labels2)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering(n_clusters=2)
labels_agg=agg.fit_predict(Xscaled)
```

```python
sns.scatterplot(x=Xscaled[:,0],y=Xscaled[:,1],c=labels_agg)
```
