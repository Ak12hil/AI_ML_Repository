```python
'''
group by species and compute
•Average sepal_length
•Maximum petal_width
•Standard deviation of sepal_width
'''
df_iris.groupby("species")["sepal_length"].mean() #avergae sepal_length
df_iris.groupby("species")["petal_width"].max() # Maximum petal_width
df_iris.groupby("species")["sepal_width"].std() #Standard deviation of sepal_width

df_iris.groupby("species").agg(
    average_sepal_length=("sepal_length","mean"),
    Maximum_petal_width=("petal_width","max"),
    Standard_deviation_of_sepal_width=("sepal_width","std"),
)
```

```python
'''
# Create a new column:
D."petal_ratio"=petal_length/petal_width
•For each species,find the average petal_ratio
'''

df_iris["petal_ratio"]=df_iris["petal_length"]/df_iris["petal_width"]
df_iris.groupby("species")["petal_ratio"].mean()
```
